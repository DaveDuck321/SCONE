```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = h5py.File('/home/tom/Downloads/jeff-3.3-hdf5/U235.h5', 'r')
```

```python
print(file['U235']["energy"].keys())
```

```python
energies_294 = np.copy(file['U235']["energy"]["294K"])
energies_900 = np.copy(file['U235']["energy"]["900K"])
energies_2500 = np.copy(file['U235']["energy"]["2500K"])
```

```python
xs_data_294 = np.copy(file['U235']['reactions']['reaction_018']['294K']['xs'])
xs_data_900 = np.copy(file['U235']['reactions']['reaction_018']['900K']['xs'])
xs_data_2500 = np.copy(file['U235']['reactions']['reaction_018']['2500K']['xs'])
```

```python
fig = plt.figure(figsize=(8, 4))
plt.loglog(energies_294, xs_data_294, label="294K")
plt.loglog(energies_2500, xs_data_2500, label="2500K")
plt.ylabel('Fission cross section (barns)')
plt.xlabel('Energy (eV)')

# Shade the resonance region
plt.axvline(1, color='C2', linestyle='--', label="Resonance")
plt.axvline(3000, color='C2', linestyle='--')
plt.legend()
```
